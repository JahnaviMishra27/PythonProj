```python
# importing required libraries
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
```

```python
# path to clean data
# reading csv into raw dataframe
df = pd.read_csv(r"C:\Users\jahna\OneDrive\Desktop\Codes\python\PythonProj\data-analysis-using-python-master\CleanData\CleanedDataSet\cleaned_autos.csv",encoding="latin-1")
```

```python
import pandas as pd

# Group by brand and vehicleType and calculate average price
df_grouped = df.groupby(['brand', 'vehicleType'], as_index=False)['price'].mean()

# Replace missing values with 0
df_grouped['price'] = df_grouped['price'].fillna(0)

# Convert price to integer
df_grouped['price'] = df_grouped['price'].astype(int)

# Rename column to avgPrice
df_grouped.rename(columns={'price': 'avgPrice'}, inplace=True)

# View first 5 rows
df_grouped.head()

```

## Average price of a vehicle by brand as well as vehicle type 

```python
tri = df_grouped.pivot(index="brand", columns="vehicleType", values="avgPrice")

# Convert values to integer
tri = tri.fillna(0).astype(int)

# Plot heatmap
plt.figure(figsize=(15,20))
sns.heatmap(tri, linewidths=1, cmap="coolwarm", annot=True, fmt="d")
plt.title("Average price of vehicles by vehicle type and brand", fontsize=18)
plt.xlabel("Vehicle Type")
plt.ylabel("Brand")
plt.show()
```
