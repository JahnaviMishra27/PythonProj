```python
# importing required libraries
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# path to clean data
# reading csv into raw dataframe
df = pd.read_csv(r"C:\Users\jahna\OneDrive\Desktop\Codes\python\PythonProj\data-analysis-using-python-master\CleanData\CleanedDataSet\cleaned_autos.csv",encoding="latin-1")
```

## Distribution of Vehicles based on Year of Registration

```python
# Distribution of vehicles based on year of registration
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(df["yearOfRegistration"], stat="density", bins=30, alpha=0.6)
sns.kdeplot(df["yearOfRegistration"], color="red")
ax.set_title('Distribution of vehicles based on Year of Registration', fontsize= 15)
plt.ylabel("Density (KDE)", fontsize= 15)
plt.xlabel("Year Of Registration", fontsize= 15)
plt.show()
```

## Variation of the price range by the vehicle type

```python
# Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x="vehicleType", y="price", data=df)

# Labels and title
plt.xlabel("Type Of Vehicle", fontsize=14)
plt.ylabel("Price Range", fontsize=14)
plt.title("Boxplot of Price based on Vehicle Type", fontsize=16)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Adjust layout so labels don’t overlap
plt.tight_layout()

plt.show()
```

## Total count of vehicles by type available on ebay for sale

```python
# Count plot to show the number of vehicles belonging to each vehicleType
plt.figure(figsize=(10,6))

# Simple countplot
ax = sns.countplot(x="vehicleType", data=df, palette="BuPu")

# Rotate labels
plt.xticks(rotation=45)
```
