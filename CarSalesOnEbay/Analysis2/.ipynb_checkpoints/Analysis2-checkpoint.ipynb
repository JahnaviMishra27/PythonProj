## No of Vehicles by Brand Available on ebay for sale

```python
# Count plot to show the number of vehicles belonging to each brand
plt.figure(figsize=(12,7))

# Simple horizontal countplot
ax = sns.countplot(y="brand", data=df, palette="Reds_r")

# Add labels and title
plt.xlabel("Count of Vehicles", fontsize=16)
plt.ylabel("Brand of Vehicles", fontsize=16)
plt.title("Count of Vehicles by Brand", fontsize=18)

# Loop through each bar in the plot
# p.get_width()  -> gives the count (length of bar) → used for x position of text
# p.get_y() + p.get_height()/2 -> finds the vertical center of the bar
# str(p.get_width()) -> converts the count to text
# va='center' -> keeps the text vertically centered

for p in ax.patches:
    plt.text(p.get_width(), p.get_y() + p.get_height()/2,
         str(p.get_width()), va='center')
```

## Average price for vehicles based on the type of vehicle as well as on the type of gearbox

```python


plt.figure(figsize=(8,5))

# Barplot with custom colors
colors = ["#00e600", "#ff8c1a", "#a180cc"]
sns.barplot(x="vehicleType", y="price", hue="gearbox", palette=colors, data=df)

# Labels and title
plt.title("Average price of vehicles by vehicle type and gearbox type", fontsize=12)
plt.xticks(rotation=45)
plt.xlabel("Type Of Vehicle", fontsize=12)
plt.ylabel("Average Price", fontsize=12)

plt.show()
```
